### Случайное разбиение на обучение и тест

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=18)
```

```python
#стандартизизируем жанные 
scaler = StandardScaler()
```

```python
X_train_scale = scaler.fit_transform(X_train)
X_valid_scale = scaler.transform(X_valid)
```

### Линейные модели

```python
#функция обучает модель, дает нашу метрику RMSE (Среднеквадратическая ошибка) и веса признаков

def train_model(model, X_train, y_train, X_valid, y_valid, feature_names):
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    
    print("RMSE: {}".format(np.sqrt(mean_squared_error((y_valid), (y_pred)))))
    
    print(pd.DataFrame(model.coef_, index=feature_names, 
                       columns=['weight']).sort_values(by='weight', ascending=False))
```

```python
linreg = LinearRegression()
```

```python
#обучим линейную регрессиию, которая решает задачу минимизации 
train_model(linreg, X_train_scale, y_train, X_valid_scale, y_valid, X_train.columns)
```

RMSE - идет отклонение на 2.29 минуты
так как у нас идет корреляция между некоторыми признаками, они имеют отрицательный вес.

```python
## применим классификаторы с регуляризацией
```

## ЛАССО - использует L1 регуляризацию

L1 - позволяет отбирать признаки, обнуляет некоторые веса, негладкий

```python
lasso = Lasso(random_state=16, alpha=0.01) #выставляем параметр альфа 
train_model(lasso, X_train_scale, y_train, X_valid_scale, y_valid, X_train.columns)
```

```python
#возможно показатель альфа 0.01 неоптимальный для регуляризации, чтобы найти оптимальный, воспользуемся классом CV, 
#который пробуем разные варианты альфа из диапазона и выбирает лучшие значения 

lasso_cv = LassoCV(random_state=16)
train_model(lasso_cv, X_train_scale, y_train, X_valid_scale, y_valid, X_train.columns)
```

В итоге, супер улучшений нет, по-прежнему высокие штрафы за мультиколиниарность 

## RIDGE- использует L2 регуляризацию

```python
ridge = Ridge(random_state=16)
train_model(ridge, X_train_scale, y_train, X_valid_scale, y_valid, X.columns)
```

```python
ridge_cv = RidgeCV()
train_model(ridge_cv, X_train_scale, y_train, X_valid_scale, y_valid, X.columns)
```
